```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

# Loading Data

```python
df_train=pd.read_csv('/kaggle/input/fraud-detection/fraudTrain.csv')
df_train.head()
```

```python
df_train.info()
```

```python
df_train.describe()
```

# Clean Data

```python
df_train.drop(columns=['Unnamed: 0','cc_num','first', 'last', 'street', 'city', 'state', 'zip', 'dob', 'trans_num','trans_date_trans_time'],inplace=True)
df_train
```

```python
df_train.isna().sum()
```

```python
df_train.columns
```

# Processing 

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
df_train["merchant"] = encoder.fit_transform(df_train["merchant"])
df_train["category"] = encoder.fit_transform(df_train["category"])
df_train["gender"] = encoder.fit_transform(df_train["gender"])
df_train["job"] = encoder.fit_transform(df_train["job"])
```

```python
df_train
```

```python
X_train=df_train.drop('is_fraud',axis=1)
y_train=df_train['is_fraud']
```

# Loading Test Data

```python
df_test=pd.read_csv('/kaggle/input/fraud-detection/fraudTest.csv')
df_test.head()
```

```python
df_test.drop(columns=['Unnamed: 0','cc_num','first', 'last', 'street', 'city', 'state', 'zip', 'dob', 'trans_num','trans_date_trans_time'],inplace=True)
df_test
```

```python
encoder = LabelEncoder()
df_test["merchant"] = encoder.fit_transform(df_test["merchant"])
df_test["category"] = encoder.fit_transform(df_test["category"])
df_test["gender"] = encoder.fit_transform(df_test["gender"])
df_test["job"] = encoder.fit_transform(df_test["job"])
```

```python
X_test=df_test.drop('is_fraud',axis=1)
y_test=df_test['is_fraud']
```

# Build Models

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,confusion_matrix
```

# Logistic Regression Model

```python
logistic_reg=LogisticRegression()
logistic_reg.fit(X_train,y_train)
```

```python
logistic_reg.score(X_train,y_train)
```

```python
logistic_reg.score(X_test,y_test)
```

```python
y_pred_reg=logistic_reg.predict(X_test)
```

```python
print(list(y_test[:30]))
print(list(y_pred_reg[:30]))
```

```python
acc_score=accuracy_score(y_test,y_pred_reg,normalize=True)

cm=confusion_matrix(y_test,y_pred_reg)
sns.heatmap(cm,cmap='RdBu')
plt.show()

print(f'confusion_matrix:\n {cm}')
print(f'accuracy_score: {acc_score}')
```

# Decision Tree Model

```python
tree=DecisionTreeClassifier( criterion='entropy',max_depth=10,min_samples_split=4)
tree.fit(X_train,y_train)
```

```python
tree.score(X_train,y_train)
```

```python
tree.score(X_test,y_test)
```

```python
y_pred_tree=tree.predict(X_test)
```

```python
print(list(y_test[:30]))
print(list(y_pred_tree[:30]))
```

```python
acc_score=accuracy_score(y_test,y_pred_tree,normalize=True)

cm=confusion_matrix(y_test,y_pred_tree)
sns.heatmap(cm,cmap='RdBu')
plt.show()

print(f'confusion_matrix:\n {cm}')
print(f'accuracy_score: {acc_score}')
```
